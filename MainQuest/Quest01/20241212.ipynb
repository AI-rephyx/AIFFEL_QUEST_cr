```python
## Q1. Account 클래스 : 은행에 가서 계좌를 개설하면 은행이름, 예금주, 계좌번호, 잔액이 설정됩니다. Account 클래스를 생성한 후 생성자(hint: 매직매서드..!!)를 구현해보세요. 생성자에서는 예금주와 초기 잔액만 입력 받습니다. 은행이름은 SC은행으로 계좌번호는 3자리-2자리-6자리 형태로 랜덤하게 생성됩니다. (은행이름: SC은행, 계좌번호: 111-11-111111)

import random

class Account: # 은행이름, 예금주, 초기 잔액
    def __init__(self, owner, balance):
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance
        self.account_number = self.aiffel_account_number()

    def aiffel_account_number(self): # 3 - 2 - 6 자리
        part1 = random.randint(100, 999)
        part2 = random.randint(10, 99)
        part3 = random.randint(100000, 999999)
        return f"{part1}-{part2}-{part3}"

    def display_info(self):
        print(f"은행이름: {self.bank}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {self.balance}원")

# 테스트
if __name__ == "__main__":
    acc1 = Account("차정은", 100000)
    acc1.display_info()

    acc2 = Account("최서윤", 500000)
    acc2.display_info()
```

```python
## Q6. 정보 출력 메서드 : Account 인스턴스에 저장된 정보를 출력하는 display_info() 메서드를 추가하세요. 잔고는 세자리마다 쉼표를 출력하세요.
## (은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)
## 세자리마다 쉼표 추가
## 계좌확인 전 예금주 선택 항목 추가

import random

class Account:
    account_count = 0  # 클래스 변수: 생성된 계좌 객체 수를 저장

    def __init__(self, owner, balance):
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance
        self.account_number = self.aiffel_account_number()

        # 계좌 생성 시 클래스 변수 증가
        Account.account_count += 1

    def aiffel_account_number(self):  # 3-2-6 자리 계좌번호 생성
        part1 = random.randint(100, 999)
        part2 = random.randint(10, 99)
        part3 = random.randint(100000, 999999)
        return f"{part1}-{part2}-{part3}"

    def deposit(self):
        try:
            amount = int(input("입금할 금액을 입력하세요: "))
            if amount < 1:
                print("입금은 최소 1원 이상만 가능합니다.")
            else:
                self.balance += amount
                print(f"{amount:,}원이 입금되었습니다. 현재 잔고는 {self.balance:,}원입니다.")
        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력해주세요.")

    def withdraw(self):
        try:
            amount = int(input("출금할 금액을 입력하세요: "))
            if amount > self.balance:
                print("잔고가 부족하여 출금할 수 없습니다.")
            elif amount < 1:
                print("출금은 최소 1원 이상만 가능합니다.")
            else:
                self.balance -= amount
                print(f"{amount:,}원이 출금되었습니다. 현재 잔고는 {self.balance:,}원입니다.")
        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력해주세요.")

    def display_info(self):  # 계좌 정보 출력
        print(f"은행이름: {self.bank}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {self.balance:,}원")

class Bank:
    def __init__(self):
        self.accounts = []  # 생성된 계좌 리스트

    def create_account(self, owner, balance):
        account = Account(owner, balance)
        self.accounts.append(account)

    def select_account(self):
        while True:
            print("\n등록된 예금주:")
            for idx, account in enumerate(self.accounts, 1):
                print(f"{idx}. {account.owner}")
            print(f"{len(self.accounts) + 1}. 종료")  # 종료 메뉴 추가
            try:
                choice = int(input("확인할 예금주의 번호를 선택하세요: "))
                if 1 <= choice <= len(self.accounts):
                    return self.accounts[choice - 1]
                elif choice == len(self.accounts) + 1:  # 종료 선택 시 프로그램 종료
                    print("프로그램을 종료합니다.")
                    return None
                else:
                    print("잘못된 번호를 선택하셨습니다. 다시 시도해주세요.")
            except ValueError:
                print("잘못된 입력입니다. 숫자를 입력해주세요.")

# 테스트
if __name__ == "__main__":
    bank = Bank()
    bank.create_account("차정은", 100000)
    bank.create_account("최서윤", 500000)

    while True:
        selected_account = bank.select_account()
        if selected_account is None:  # 프로그램 종료 선택
            break

        print("\n1: 입금 | 2: 출금 | 3: 계좌 정보 출력 | 4: 종료")
        try:
            choice = int(input(f"{selected_account.owner}님의 계좌에서 실행할 작업 번호를 선택하세요: "))
            if choice == 1:
                selected_account.deposit()
            elif choice == 2:
                selected_account.withdraw()
            elif choice == 3:
                selected_account.display_info()
            elif choice == 4:
                print("작업을 종료하고 예금주 선택 메뉴로 돌아갑니다.")
            else:
                print("잘못된 번호를 입력하셨습니다. 다시 선택해주세요.")
        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력해주세요.")
```

```python
## Q7. 이자 지급하기 : 입금 횟수가 5회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가되도록 코드를 변경해보세요.
## 예금주 선택으로 돌아가지 않고 입금, 출금, 계좌 정보 출력을 반복할 수 있는 기능 추가

import random

class Account:
    account_count = 0  # 클래스 변수: 생성된 계좌 객체 수를 저장

    def __init__(self, owner, balance):
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance
        self.account_number = self.aiffel_account_number()
        self.deposit_count = 0  # 입금 횟수 초기화

        # 계좌 생성 시 클래스 변수 증가
        Account.account_count += 1

    def aiffel_account_number(self):  # 3-2-6 자리 계좌번호 생성
        part1 = random.randint(100, 999)
        part2 = random.randint(10, 99)
        part3 = random.randint(100000, 999999)
        return f"{part1}-{part2}-{part3}"

    def deposit(self):
        try:
            amount = int(input("입금할 금액을 입력하세요: "))
            if amount < 1:
                print("입금은 최소 1원 이상만 가능합니다.")
            else:
                self.balance += amount
                self.deposit_count += 1  # 입금 횟수 증가
                print(f"{amount:,}원이 입금되었습니다. 현재 잔고는 {self.balance:,}원입니다.")

                # 입금 횟수 5회마다 이자 지급
                if self.deposit_count == 5:
                    interest = int(self.balance * 0.01)
                    self.balance += interest
                    self.deposit_count = 0  # 입금 횟수 초기화
                    print(f"축하합니다! 잔고의 1%({interest:,}원)가 이자로 지급되었습니다. 현재 잔고는 {self.balance:,}원입니다.")
        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력해주세요.")

    def withdraw(self):
        try:
            amount = int(input("출금할 금액을 입력하세요: "))
            if amount > self.balance:
                print("잔고가 부족하여 출금할 수 없습니다.")
            elif amount < 1:
                print("출금은 최소 1원 이상만 가능합니다.")
            else:
                self.balance -= amount
                print(f"{amount:,}원이 출금되었습니다. 현재 잔고는 {self.balance:,}원입니다.")
        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력해주세요.")

    def display_info(self):  # 계좌 정보 출력
        print(f"은행이름: {self.bank}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {self.balance:,}원")

class Bank:
    def __init__(self):
        self.accounts = []  # 생성된 계좌 리스트

    def create_account(self, owner, balance):
        account = Account(owner, balance)
        self.accounts.append(account)

    def select_account(self):
        while True:
            print("\n등록된 예금주:")
            for idx, account in enumerate(self.accounts, 1):
                print(f"{idx}. {account.owner}")
            print(f"{len(self.accounts) + 1}. 종료")  # 종료 메뉴 추가
            try:
                choice = int(input("확인할 예금주의 번호를 선택하세요: "))
                if 1 <= choice <= len(self.accounts):
                    return self.accounts[choice - 1]
                elif choice == len(self.accounts) + 1:  # 종료 선택 시 프로그램 종료
                    print("프로그램을 종료합니다.")
                    return None
                else:
                    print("잘못된 번호를 선택하셨습니다. 다시 시도해주세요.")
            except ValueError:
                print("잘못된 입력입니다. 숫자를 입력해주세요.")

# 테스트
if __name__ == "__main__":
    bank = Bank()
    bank.create_account("차정은", 100000)
    bank.create_account("최서윤", 500000)

    while True:
        selected_account = bank.select_account()
        if selected_account is None:  # 프로그램 종료 선택
            break

        while True:
            print("\n1: 입금 | 2: 출금 | 3: 계좌 정보 출력 | 4: 초기 메뉴로 돌아가기")
            try:
                choice = int(input(f"{selected_account.owner}님의 계좌에서 실행할 작업 번호를 선택하세요: "))
                if choice == 1:
                    selected_account.deposit()
                elif choice == 2:
                    selected_account.withdraw()
                elif choice == 3:
                    selected_account.display_info()
                elif choice == 4:
                    print("초기 메뉴로 돌아갑니다.")
                    break  # 초기 메뉴로 돌아가기
                else:
                    print("잘못된 번호를 입력하셨습니다. 다시 선택해주세요.")
            except ValueError:
                print("잘못된 입력입니다. 숫자를 입력해주세요.")
```

```python
## Q10. 입금과 출금 내역이 기록되도록 코드를 업데이트 하세요.
## (입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드를 추가하세요)

import random

class Account:
    account_count = 0  # 클래스 변수: 생성된 계좌 객체 수를 저장

    def __init__(self, owner, balance):
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance
        self.account_number = self.aiffel_account_number()
        self.deposit_count = 0  # 입금 횟수 초기화
        self.deposit_records = []  # 입금 내역
        self.withdraw_records = []  # 출금 내역

        # 계좌 생성 시 클래스 변수 증가
        Account.account_count += 1

    def aiffel_account_number(self):  # 3-2-6 자리 계좌번호 생성
        part1 = random.randint(100, 999)
        part2 = random.randint(10, 99)
        part3 = random.randint(100000, 999999)
        return f"{part1}-{part2}-{part3}"

    def deposit(self):
        try:
            amount = int(input("입금할 금액을 입력하세요: "))
            if amount < 1:
                print("입금은 최소 1원 이상만 가능합니다.")
            else:
                self.balance += amount
                self.deposit_count += 1  # 입금 횟수 증가
                self.deposit_records.append(amount)  # 입금 내역 기록
                print(f"{amount:,}원이 입금되었습니다. 현재 잔고는 {self.balance:,}원입니다.")

                # 입금 횟수 5회마다 이자 지급
                if self.deposit_count == 5:
                    interest = int(self.balance * 0.01)
                    self.balance += interest
                    self.deposit_count = 0  # 입금 횟수 초기화
                    print(f"축하합니다! 잔고의 1%({interest:,}원)가 이자로 지급되었습니다. 현재 잔고는 {self.balance:,}원입니다.")
        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력해주세요.")

    def withdraw(self):
        try:
            amount = int(input("출금할 금액을 입력하세요: "))
            if amount > self.balance:
                print("잔고가 부족하여 출금할 수 없습니다.")
            elif amount < 1:
                print("출금은 최소 1원 이상만 가능합니다.")
            else:
                self.balance -= amount
                self.withdraw_records.append(amount)  # 출금 내역 기록
                print(f"{amount:,}원이 출금되었습니다. 현재 잔고는 {self.balance:,}원입니다.")
        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력해주세요.")

    def deposit_history(self):
        print(f"\n{self.owner}님의 입금 내역:")
        if not self.deposit_records:
            print("입금 내역이 없습니다.")
        else:
            for idx, amount in enumerate(self.deposit_records, 1):
                print(f"{idx}. {amount:,}원")

    def withdraw_history(self):
        print(f"\n{self.owner}님의 출금 내역:")
        if not self.withdraw_records:
            print("출금 내역이 없습니다.")
        else:
            for idx, amount in enumerate(self.withdraw_records, 1):
                print(f"{idx}. {amount:,}원")

    def display_info(self):  # 계좌 정보 출력
        print(f"은행이름: {self.bank}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {self.balance:,}원")

class Bank:
    def __init__(self):
        self.accounts = []  # 생성된 계좌 리스트

    def create_account(self, owner, balance):
        account = Account(owner, balance)
        self.accounts.append(account)

    def select_account(self):
        while True:
            print("\n등록된 예금주:")
            for idx, account in enumerate(self.accounts, 1):
                print(f"{idx}. {account.owner}")
            print(f"{len(self.accounts) + 1}. 관리자")  # 관리자 메뉴 추가
            print(f"{len(self.accounts) + 2}. 종료")  # 종료 메뉴 추가
            try:
                choice = int(input("확인할 항목의 번호를 선택하세요: "))
                if 1 <= choice <= len(self.accounts):  # 예금주 선택
                    return self.accounts[choice - 1]
                elif choice == len(self.accounts) + 1:  # 관리자 선택
                    self.admin_menu()
                elif choice == len(self.accounts) + 2:  # 종료 선택
                    print("프로그램을 종료합니다.")
                    return None
                else:
                    print("잘못된 번호를 선택하셨습니다. 다시 시도해주세요.")
            except ValueError:
                print("잘못된 입력입니다. 숫자를 입력해주세요.")

    def admin_menu(self):
        print("\n관리자 메뉴")
        while True:
            print("1: 잔고 100만원 이상 고객 확인 | 2: 이전 메뉴로 돌아가기")
            try:
                admin_choice = int(input("원하는 작업의 번호를 입력하세요: "))
                if admin_choice == 1:
                    self.display_high_balance_accounts()
                elif admin_choice == 2:
                    print("이전 메뉴로 돌아갑니다.")
                    break
                else:
                    print("잘못된 번호를 입력하셨습니다. 다시 선택해주세요.")
            except ValueError:
                print("잘못된 입력입니다. 숫자를 입력해주세요.")

    def display_high_balance_accounts(self):
        print("\n잔고가 100만원 이상인 고객 정보:")
        for account in self.accounts:
            if account.balance >= 1000000:
                account.display_info()

# 테스트
if __name__ == "__main__":
    bank = Bank()

    # 여러 객체 생성 및 리스트에 저장
    bank.create_account("차정은", 100000)
    bank.create_account("최서윤", 500000)
    bank.create_account("양건우", 1000000)
    bank.create_account("일론 머스크", 1000000000)

    while True:
        selected_account = bank.select_account()
        if selected_account is None:  # 프로그램 종료 선택
            break

        while True:
            print("\n1: 입금 | 2: 출금 | 3: 계좌 정보 출력 | 4: 입금 내역 | 5: 출금 내역 | 6: 초기 메뉴로 돌아가기")
            try:
                choice = int(input(f"{selected_account.owner}님의 계좌에서 실행할 작업 번호를 선택하세요: "))
                if choice == 1:
                    selected_account.deposit()
                elif choice == 2:
                    selected_account.withdraw()
                elif choice == 3:
                    selected_account.display_info()
                elif choice == 4:
                    selected_account.deposit_history()
                elif choice == 5:
                    selected_account.withdraw_history()
                elif choice == 6:
                    print("초기 메뉴로 돌아갑니다.")
                    break  # 초기 메뉴로 돌아가기
                else:
                    print("잘못된 번호를 입력하셨습니다. 다시 선택해주세요.")
            except ValueError:
                print("잘못된 입력입니다. 숫자를 입력해주세요.")
```

```python
## Q1. Character 클래스 만들기

import random

class Character:
    def __init__(self, name, level, health, attack_power, defense):
        """
        Character 클래스의 생성자

        :param name: 캐릭터 이름
        :param level: 캐릭터 레벨
        :param health: 캐릭터 최대 체력
        :param attack_power: 캐릭터 공격력
        :param defense: 캐릭터 방어력
        """
        self.name = name
        self.level = level
        self.health = health
        self.current_health = health  # 현재 체력
        self.attack_power = attack_power
        self.defense = defense

    def is_alive(self):
        """
        캐릭터가 살아있는지 확인하는 메서드

        :return: 현재 체력이 0 이상이면 True, 아니면 False
        """
        return self.current_health > 0

    def take_damage(self, damage):
        """
        캐릭터가 피해를 받는 메서드

        :param damage: 받은 데미지
        """
        net_damage = max(0, damage - self.defense)  # 방어력을 제외한 실질 데미지
        self.current_health = max(0, self.current_health - net_damage)  # 체력 감소
        print(f"{self.name}이(가) {net_damage}의 피해를 입었습니다. 현재 체력: {self.current_health}/{self.health}")

    def attack_target(self, target):
        """
        타겟에게 데미지를 입히는 메서드

        :param target: 공격받는 대상 (Character 인스턴스)
        """
        damage = random.randint(1, self.attack_power)  # 1부터 공격력 사이의 랜덤 데미지
        print(f"{self.name}이(가) {target.name}에게 {damage}의 데미지를 입혔습니다.")
        target.take_damage(damage)


# 테스트 코드
if __name__ == "__main__":
    # 사용자 입력으로 플레이어 이름 설정
    player_name = input("플레이어 이름을 입력하세요: ")
    print()  # 줄바꿈
    print("딴딴- 모험이 시작되었습니다!")
    print()  # 줄바꿈

    # 플레이어와 몬스터 생성
    player = Character(player_name, 1, 100, 20, 5)
    monster = Character("Monster", 1, 50, 15, 3)

    # 전투 시뮬레이션
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)
        else:
            print(f"{monster.name}이(가) 쓰러졌습니다!")

    if player.is_alive():
        print(f"{player.name}이(가) 승리했습니다!")
    else:
        print(f"{player.name}이(가) 쓰러졌습니다.")
```

```python
## Q2. Player 클래스와 Monster 클래스 만들기
## 죽거나 레벨업 할 때까지 전투 반복 (레벨업 후 전투 한번 더 진행) 기능 추가
## 레벨 1에는 몬스터 레벨 제한이 있어 수월한 플레이가 가능하도록 밸런스 패치
## 전투 3번 반복 후 체력 회복을 할 수 있는 이벤트 추가

import random

class Character:
    def __init__(self, name, level, health, attack_power, defense):
        self.name = name
        self.level = level
        self.health = health
        self.current_health = health
        self.attack_power = attack_power
        self.defense = defense

    def is_alive(self):
        return self.current_health > 0

    def take_damage(self, damage):
        net_damage = max(0, damage - self.defense)
        self.current_health = max(0, self.current_health - net_damage)
        print(f"{self.name}이(가) {net_damage}의 피해를 입었습니다. 현재 체력: {self.current_health}/{self.health}")

    def attack_target(self, target):
        damage = random.randint(1, self.attack_power)
        print(f"{self.name}이(가) {target.name}에게 {damage}의 데미지를 입혔습니다.")
        target.take_damage(damage)


class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, health=100, attack_power=25, defense=5)
        self.experience = 0
        self.level_up_bonus = False  # 레벨업 후 전투 1회 체크

    def gain_experience(self, exp):
        self.experience += exp
        print(f"{self.name}이(가) {exp} 경험치를 획득했습니다. 현재 경험치: {self.experience}")
        self.level_up()

    def level_up(self):
        if self.experience >= 50:
            self.level += 1
            self.attack_power += 10
            self.defense += 5
            self.experience -= 50
            self.level_up_bonus = True  # 레벨업 후 전투 1회 트리거
            print(f"{self.name}이(가) 레벨업 했습니다! 레벨: {self.level}, 공격력: {self.attack_power}, 방어력: {self.defense}")

    def restore_health(self):
        self.current_health = self.health
        print(f"{self.name}이(가) 안식의 정원에서 체력을 완전히 회복했습니다! 현재 체력: {self.current_health}/{self.health}")


class Monster(Character):
    def __init__(self, name, level):
        health = random.randint(10, 30) * level
        attack_power = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, health, attack_power, defense)


if __name__ == "__main__":
    player_name = input("플레이어 이름을 입력하세요: ")
    print("\n딴딴- 모험이 시작되었습니다!\n")
    player = Player(player_name)
    battle_count = 0

    while player.is_alive():
        # 몬스터 생성
        if player.level == 1:
            monster_level = random.randint(1, 1)  # 레벨 1에는 몬스터 레벨 최대 1까지
        else:
            monster_level = random.randint(1, player.level + 2)  # 레벨 2 이상에서는 플레이어 레벨 +2까지

        monster = Monster(f"Monster (레벨 {monster_level})", monster_level)
        print(f"\n{monster.name}이(가) 등장했습니다!\n")

        # 전투
        while player.is_alive() and monster.is_alive():
            player.attack_target(monster)
            if monster.is_alive():
                monster.attack_target(player)
            else:
                print(f"{monster.name}이(가) 쓰러졌습니다!")
                player.gain_experience(20)  # 경험치 획득

        # 플레이어 죽음 확인
        if not player.is_alive():
            print(f"{player.name}이(가) 쓰러졌습니다. 게임이 종료됩니다.")
            break

        battle_count += 1

        # 안식의 정원 이벤트
        if battle_count == 3:
            print("\n안식의 정원이 나타났습니다. 체력을 회복합니다.")
            player.restore_health()
            battle_count = 0  # 전투 카운트 초기화
            print("\n하루가 지났습니다.\n")

        # 레벨업 후 1회 전투 후 종료
        if player.level_up_bonus:
            print("\n모험을 지속합니다.")
            monster = Monster(f"Monster (레벨 {player.level})", player.level)
            print(f"\n{monster.name}이(가) 등장했습니다!\n")
            while player.is_alive() and monster.is_alive():
                player.attack_target(monster)
                if monster.is_alive():
                    monster.attack_target(player)
                else:
                    print(f"{monster.name}이(가) 쓰러졌습니다!")
            print("\n추가 전투가 종료되었습니다. 게임을 종료합니다.")
            break
```

```python
## Q3. battle 함수 만들기

import random

class Character:
    def __init__(self, name, level, health, attack_power, defense):
        self.name = name
        self.level = level
        self.health = health
        self.current_health = health
        self.attack_power = attack_power
        self.defense = defense

    def is_alive(self):
        return self.current_health > 0

    def take_damage(self, damage):
        net_damage = max(0, damage - self.defense)
        self.current_health = max(0, self.current_health - net_damage)
        print(f"{self.name}이(가) {net_damage}의 피해를 입었습니다. 현재 체력: {self.current_health}/{self.health}")

    def attack_target(self, target):
        damage = random.randint(1, self.attack_power)
        print(f"{self.name}이(가) {target.name}에게 {damage}의 데미지를 입혔습니다.")
        target.take_damage(damage)


class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, health=100, attack_power=25, defense=5)
        self.experience = 0

    def gain_experience(self, exp):
        self.experience += exp
        print(f"{self.name}이(가) {exp} 경험치를 획득했습니다. 현재 경험치: {self.experience}")
        self.level_up()

    def level_up(self):
        if self.experience >= 50:
            self.level += 1
            self.attack_power += 10
            self.defense += 5
            self.experience -= 50
            print(f"{self.name}이(가) 레벨업 했습니다! 레벨: {self.level}, 공격력: {self.attack_power}, 방어력: {self.defense}")

    def restore_health(self):
        self.current_health = self.health
        print(f"{self.name}이(가) 안식의 정원에서 체력을 완전히 회복했습니다! 현재 체력: {self.current_health}/{self.health}")


class Monster(Character):
    def __init__(self, name, level):
        health = random.randint(10, 30) * level
        attack_power = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, health, attack_power, defense)


def battle(player, monster):
    print(f"\n{monster.name}이(가) 등장했습니다!\n")

    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)

    if player.is_alive():
        print("전투 승리!")
        player.gain_experience(monster.level * 20)
    else:
        print("전투 패배..")


if __name__ == "__main__":
    player_name = input("플레이어 이름을 입력하세요: ")
    print("\n딴딴- 모험이 시작되었습니다!\n")
    player = Player(player_name)
    battle_count = 0

    while player.is_alive():
        # 몬스터 생성
        if player.level == 1:
            monster_level = random.randint(1, 1)  # 레벨 1에는 몬스터 레벨 최대 1까지
        else:
            monster_level = random.randint(1, player.level + 2)  # 레벨 2 이상에서는 플레이어 레벨 +2까지

        monster = Monster(f"Monster (레벨 {monster_level})", monster_level)

        # 전투
        battle(player, monster)

        # 플레이어 죽음 확인
        if not player.is_alive():
            print(f"{player.name}이(가) 쓰러졌습니다. 게임이 종료됩니다.")
            break

        battle_count += 1

        # 안식의 정원 이벤트
        if battle_count == 3:
            print("\n안식의 정원이 나타났습니다. 체력을 회복합니다.")
            player.restore_health()
            battle_count = 0  # 전투 카운트 초기화
            print("\n하루가 지났습니다.\n")
```

```python
## Q4. main 함수 만들기
## 레벨에 따라 다른 몬스터가 등장하게끔 패치
## 퀘스트가 끝나는 것이 아쉬워 모든 몬스터를 처치한 후 기믹 추가

import random
import time

class Character:
    def __init__(self, name, level, health, attack_power, defense):
        self.name = name
        self.level = level
        self.health = health
        self.current_health = health
        self.attack_power = attack_power
        self.defense = defense

    def is_alive(self):
        return self.current_health > 0

    def take_damage(self, damage):
        net_damage = max(0, damage - self.defense)
        self.current_health = max(0, self.current_health - net_damage)
        print(f"{self.name}이(가) {net_damage}의 피해를 입었습니다. 현재 체력: {self.current_health}/{self.health}")

    def attack_target(self, target):
        damage = random.randint(1, self.attack_power)
        print(f"{self.name}이(가) {target.name}에게 {damage}의 데미지를 입혔습니다.")
        target.take_damage(damage)


class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, health=100, attack_power=25, defense=5)
        self.experience = 0

    def gain_experience(self, exp):
        self.experience += exp
        print(f"{self.name}이(가) {exp} 경험치를 획득했습니다. 현재 경험치: {self.experience}")
        self.level_up()

    def level_up(self):
        if self.experience >= 50:
            self.level += 1
            self.attack_power += 10
            self.defense += 5
            self.experience -= 50
            print(f"{self.name}이(가) 레벨업 했습니다! 레벨: {self.level}, 공격력: {self.attack_power}, 방어력: {self.defense}")

    def restore_health(self):
        self.current_health = self.health
        print(f"{self.name}이(가) 안식의 정원에서 체력을 완전히 회복했습니다! 현재 체력: {self.current_health}/{self.health}")


class Monster(Character):
    def __init__(self, name, level):
        health = random.randint(10, 30) * level
        attack_power = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, health, attack_power, defense)


def battle(player, monster):
    print(f"\n{monster.name}이(가) 등장했습니다!\n")
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)

    if player.is_alive():
        print("전투 승리!")
        player.gain_experience(monster.level * 20)
    else:
        print("전투 패배..")


def main():
    monster_dict = {
        "일반 몬스터": 1,
        "슬라임": 2,
        "고블린": 3,
        "오크": 4
    }

    player_name = input("플레이어 이름을 입력하세요: ")
    print("\n딴딴- 모험이 시작되었습니다!\n")
    player = Player(player_name)
    defeated_monsters = {"슬라임": False, "고블린": False, "오크": False}
    battle_count = 0

    while player.is_alive():
        # 플레이어 레벨에 따라 등장 몬스터 필터링
        if player.level == 1:
            available_monsters = {"일반 몬스터": 1}
        elif player.level == 2:
            available_monsters = {name: level for name, level in monster_dict.items() if level <= 2}
        else:
            available_monsters = {name: level for name, level in monster_dict.items() if level <= player.level}

        monster_name, monster_level = random.choice(list(available_monsters.items()))
        monster = Monster(monster_name, monster_level)

        battle(player, monster)
        battle_count += 1

        if monster_name in defeated_monsters:
            defeated_monsters[monster_name] = True

        if not player.is_alive():
            print(f"\n{player.name}이(가) 쓰러졌습니다. 게임오버.")
            return

        # 안식의 정원 이벤트
        if battle_count == 3:
            print("\n안식의 정원이 나타났습니다. 체력을 회복합니다.")
            player.restore_health()
            battle_count = 0
            print("\n하루가 지났습니다.\n")

        # 모든 특별 몬스터 처치 시 이벤트 발생
        if all(defeated_monsters.values()):
            print(f"\n축하합니다! {player.name}이(가) 모든 몬스터를 물리쳤습니다!\n")
            print("당신은 시대가 기억하는 영웅이 되었습니다.")

            while True:
                choice = input("\n모험을 지속하시겠습니까? (y/n): ").lower()
                if choice == 'y':
                    print("\n지금까지 데모 버전을 플레이해주셔서 감사합니다.")
                    print("\n1. 모험을 이어가기 위하여 10$ 결제하기\n2. 모험을 마치기")
                    try:
                        sub_choice = int(input("선택하세요: "))
                        if sub_choice == 1:
                            print("\n1002 047 721180 우리은행 양건우")
                            print("3...")
                            time.sleep(1)
                            print("2...")
                            time.sleep(1)
                            print("1...")
                            time.sleep(1)
                            print("\n땡 시간이 초과되었습니다.")
                            print(f"\n영웅 {player.name}의 망설임으로 인해 세상은 몬스터에게 멸망당했습니다.\nGame over.")
                            return
                        elif sub_choice == 2:
                            print(f"\n영웅 {player.name}의 모험은 여기서 막을 내렸습니다. 영웅의 활동이 중단되자 세상은 몬스터로 인해 멸망했습니다.\nGame over.")
                            return
                        else:
                            print("잘못된 입력입니다. 다시 입력해주세요.")
                    except ValueError:
                        print("잘못된 입력입니다. 숫자를 입력해주세요.")
                elif choice == 'n':
                    print(f"\n영웅 {player.name}의 모험은 여기서 막을 내렸습니다. 영웅의 활동이 중단되자 세상은 몬스터로 인해 멸망했습니다.\nGame over.")
                    return
                else:
                    print("잘못된 입력입니다. 다시 입력해주세요.")


if __name__ == "__main__":
    main()
```
